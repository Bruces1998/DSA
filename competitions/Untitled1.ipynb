```python
def get_sum(string, n):
    if n == 1:
        return (-1, -1)
    for i in range(n-1):
        summ = 0
        for j in range(i, n):
            if string[j] == "a":
                summ+=1
            else:
                summ -=1
                
            if summ == 0:
                return (i+1, j+1)
            
    return (-1, -1)

T = int(input())
inp_n = []
inp_s = []
for _ in range(T):
    inp_n.append(int(input()))
    inp_s.append(input())
    
for k in range(T):
    ans = get_sum(inp_s[k], inp_n[k])
    print(ans[0], ans[1])
```

```python
from itertools import permutations
def is_nice(summ, permutation, arr, n):
    i = 0
    last = -1
    while(summ!=0):
        print(i, last, arr)
        if arr[permutation[i]]==0:
            i = (i+1)%n
            continue
            
        else:
            if last == i:
                return False
            
            else:
                summ -= 1
                arr[permutation[i]] -= 1
                last = i
                i = (i+1)%n
                
    return True
def get_result(n, arr):
    summa = sum(arr)
    if summa == 0:
        return 0
    permutat = list(permutations(range(0, n)))
    count = 0
    for perm in permutat:
        if is_nice(summa, perm, arr, n):
            count+=1
            print(count)
            
            
    return count%998244353
    
    
T = int(input())
inp_n = []
inp_s = []
for _ in range(T):
    inp_n.append(int(input()))
    inp_s.append(list(map(int, input().split(" "))))
    
    
for k in range(T):
    ans = get_result( inp_n[k], inp_s[k])
    print(ans)
    
```
